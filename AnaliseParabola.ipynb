```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize
```

```python
# Time Serie:
x = np.array([1, 2, 3, 4 ,5])
y = np.array([10, 23, 26, 25, 20])

plt.plot(x,y)
plt.ylim(0,30)
plt.show()
```

```python
def f(x,b0,b1, b2):
  y = b0 + b1 * x + b2 * (x**2)
  return y
```

```python
guess = [1,1,1]

optimize.curve_fit(f,x,y,guess)

params, params_covariance = optimize.curve_fit(f,x,y,guess)
```

```python
b0 = params[0]
b1 = params[1]
b2= params[2]

print("b0 =", b0)
print("b1 =", b1)
print("b2 =", b2)
```

```python
# Calculando a métrica do modelo - SSE
y_fit = f(x,b0,b1,b2)
MSE = sum((y-y_fit)**2)/len(y)

# Plotando os pontos, a reta ajustada e a métrica MSE
plt.plot(x, y_fit, color="red", label="Linear Fitting")
plt.plot(x,y, color="blue", label="Time Serie")
plt.legend(loc='best')
plt.ylim(0,30)
print("MSE =", MSE)

# Print dos parâmetros
print("b0 =", b0.round(2))
print("b1 =", b1.round(2))
print("b2 =", b2.round(2))
```

```python
# Próximo período futuro: x = 6
y_forecast = f(6, b0, b1, b2)
print(y_forecast.round(2))
```
